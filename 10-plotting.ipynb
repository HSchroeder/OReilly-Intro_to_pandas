```python
import seaborn as sns
import matplotlib.pyplot as plt
# import plotnine as p9
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
```

```python
sns.lmplot(x="total_bill", y="tip", hue="time", data=tips, fit_reg=False, col="day")
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, fit_reg=False, col="day", row="time")
plt.show()
```
