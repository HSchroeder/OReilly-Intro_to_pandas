```python
import seaborn as sns
import matplotlib.pyplot as plt
# import plotnine as p9
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show() #optional?
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="time", fit_reg=False, col="day")
plt.show() #optional?
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="time", fit_reg=False, col="day", row="time")
plt.show() #optional?
```
